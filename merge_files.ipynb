```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('./data/Prod_rev1.xlsx')
df1
```

```python
df2 = pd.read_excel('./data/March2024.xlsx')
df2
```

### Ton

```python
df1['Ton']
```

```python
df2['Ton']
```

```python
# remove subfix 'Ton(T)' from df1's Ton and changed both dfs' Ton to float
df1['Ton'] = df1['Ton'].str.removesuffix(' Ton(T)').astype(float)
df2['Ton'] = df2['Ton'].astype(float)

df1['Ton'].dtype == df2['Ton'].dtype
```

### Unnamed: 200

```python
df1['Unnamed: 200']
```

```python
df2['Unnamed: 200']
```

### Merge Files

```python
merged_df = pd.concat([df1, df2]).drop_duplicates().reset_index(drop=True).sort_values(by='Date')
merged_df
```

```python
org_df = pd.read_excel('./data/Prod.xlsx')
org_df
```

```python
not_in_org = merged_df.columns.difference(org_df.columns)
not_in_org
```

```python
merged_df.drop(columns=not_in_org, inplace=True)
merged_df.shape
```

```python
not_in_merged = org_df.columns.difference(merged_df.columns)
not_in_merged
```

```python
org_df[not_in_merged]
```

```python
merged_df[['0.18', '0.19', '0.20', '0.21']] = 0
merged_df['Unnamed: 194'] = None
merged_df.shape
```

```python
merged_df = merged_df.reindex(org_df.columns, axis='columns')
```

```python
(merged_df.dtypes == org_df.dtypes)[(merged_df.dtypes == org_df.dtypes) == False]
```

```python
merged_df['Date'] = pd.to_datetime(merged_df['Date'], format='%Y-%m-%d')
merged_df['Date']
```

```python
for col in merged_df.columns:
    try:
        new_col = merged_df[col].replace("", 0).replace("-", 0).astype(float)
        merged_df[col] = new_col
    except:
        print(f"'{col}' is not convertible to Float")
```

```python
from pandas.api.types import is_numeric_dtype

for col in merged_df.columns:
    if is_numeric_dtype(merged_df[col]):
        new_col = merged_df[col].replace(np.NaN, 0)
        merged_df[col] = new_col
```

```python
for col in merged_df.columns:
    try:
        new_col = merged_df[col].astype(org_df[col].dtype)
        merged_df[col] = new_col
    except:
        print(f"'{col}' is not convertible to that of the original")
```

```python
with pd.ExcelWriter('./data/Merged_Prod.xlsx', engine='xlsxwriter', datetime_format='m/d/yyyy') as writer:
    merged_df.to_excel(writer, index=False)
```
